```python
full_list = []
name_list = []

for i in stock_list:
  try:
    df = yf.download(i, start="2013-01-01", end="2023-01-01")
    df = df.reset_index()
    df['return'] = df['Close'].pct_change()
    df = df.dropna()

    name_list.append(i)

    return_list = df['return'].to_list()

    full_list.append(return_list)

  except Exception:
    pass
```

```python
df = pd.DataFrame(full_list)
df = df.T
df.columns = name_list
df = df.dropna(axis=1)
df.head()
```

```python
df.shape
```

```python
random_tickers = []

for i in range(20):
  random_tickers.append(df.columns[i])
```

```python
f_list = []

for i in random_tickers:
  m_df = df[i]

  for i in range(2496):
    t = m_df[(i+1):(i+22)].to_numpy()
    f_list.append(t)
```

```python
full_array = np.array([f_list])
```

```python
full_array
```

```python
final_df = pd.DataFrame(full_array[0])
final_df
```

```python
shuffled_df = final_df.sample(n=len(final_df))
shuffled_df = shuffled_df.reset_index(drop=True)
shuffled_df
```
